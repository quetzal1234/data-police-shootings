```python
import pandas as pd
import bqplot
import numpy as np
import traitlets
import ipywidgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("fatal-police-shootings-data.csv")
df.head()
```

```python
race = df.groupby("race")["id"].count()
#race = race.reset_index()
race
```

```python
mentalillness = df.groupby(["race", "signs_of_mental_illness"])["id"].count()
#mentalillness = mentalillness.reset_index()
mentalillness
```

```python
bodycamera = df.groupby(["race", "body_camera"])["id"].count()
bodycamera = bodycamera.reset_index()
bodycamera
```

```python
#Basic histogram
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.LinearScale()

x_ax = bqplot.Axis(scale = x_sc)
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical')

hist = bqplot.Hist(sample = race,
                   normalized = False,
                    scales = {'sample': x_sc, 'count': y_sc},
                    bins = 6)
fig = bqplot.Figure(marks = [hist], axes = [x_ax, y_ax])
fig
```
